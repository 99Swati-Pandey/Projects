```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Nishant\\Machine Learning\\Datasets\\Admission_Predict.csv")
df.head()
```

```python
df1 = df.drop(['Serial_No'],axis='columns')
df1.head()
```

```python
X = df1.drop(['Chance_Admit'],axis='columns')
X.head()
```

```python
y = df1['Chance_Admit']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
